# Clase 4: _Support Vector Machines_ y _Kernelization_

### Mapa de Características (Feature Maps)
* Recordemos los modelos lineales

$$\hat{y} = \mathbf{w}\mathbf{x} = \sum_{i=0}^{p} w_i \cdot x_i = w_0 \cdot x_0 + w_1 \cdot x_1 + ... + w_p \cdot x_p $$

* Cuando no podemos ajustar bien los datos con modelos lineales, podemos aprender modelos más complejos simplemente agregando más dimensiones
* Mapa de características (o *basis expansion*) $\phi$: $ X \rightarrow \mathbb{R}^d $

$$y=\textbf{w}^T\textbf{x} \rightarrow  y=\textbf{w}^T\phi(\textbf{x})$$

* Es posible que aún necesite MUCHAS dimensiones para adaptarse a los datos
    * Memoria y costo computacional
    * Más probable que se caiga en overfitting


Ejemplo: regresión Ridge

````python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

X, y = mglearn.datasets.make_wave(n_samples=100)
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1, 1)

reg = Ridge().fit(X, y)
print("Coefficients:",reg.coef_)
plt.rcParams['figure.figsize'] = [20, 10]
plt.plot(line, reg.predict(line), label="linear regression")

plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best");
````

* Agregue todos los polinomios $x^d$ hasta el grado D. ¿Qué tan grande debe ser D?
* También podemos calcular todos los polinomios y todas las interacciones entre características (e.j. $x \cdot x^2$). Esto lleva a D$^2$ características.

````python
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures

# include polynomials up to x ** 10:
# the default "include_bias=True" adds a feature that's constantly 1
poly = PolynomialFeatures(degree=10, include_bias=False)
poly.fit(X)
X_poly = poly.transform(X)
styles = [dict(selector="td", props=[("font-size", "80%")]),dict(selector="th", props=[("font-size", "80%")])]
pd.DataFrame(X_poly, columns=poly.get_feature_names()).head().style.set_table_styles(styles)
````

Ajustar Ridge nuevamente:

````python
reg = Ridge().fit(X_poly, y)
print("Coefficients:",reg.coef_)
line_poly = poly.transform(line)
plt.plot(line, reg.predict(line_poly), label='polynomial linear regression')
plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best");
````


* Ridge tiene una solución de forma cerrada que podemos calcular con álgebra lineal:
$$w^{*} = (X^{T}X + \lambda I)^{-1} X^T Y$$
* Dado que X tiene $n$ filas (ejemplos) y $d$ columnas (características), $X^{T}X$ tiene dimensionalidad $d x d$
* Por lo tanto, Ridge es cuadrática en la cantidad de características, $\mathcal{O}(d^2n)$
* Después del mapa de características $\Phi$, obtenemnos
$$w^{*} = (\Phi(X)^{T}\Phi(X) + \lambda I)^{-1} \Phi(X)^T Y$$
* Dado que $\Phi$ aumenta $d$ un *montón*, $\Phi(X)^{T}\Phi(X)$ se vuelve _enorme_
* Esta discusión la continuaremos más adelante ...

### Support vector machines o Máquinas de soporte vectorial (SVMs)

- En varios otros modelos lineales, minimizamos el error de clasificación errónea
- En SVMs, El objetivo de optimización es maximizar el _margin_ (margen)
- El __margin__ es la distancia entre el hiperplano de separación y los *vectores de soporte*
- Los __vectores de soporte__ son las muestras de entrenamiento más cercanas al hiperplano
- Intuición: los márgenes grandes generalizan mejor, los márgenes pequeños pueden ser propensos al sobreajuste


![](https://drive.google.com/uc?id=1pIe-1Zg2EOKKhgjv3fPCp7BD0250iyPj)

#### Margen máximo

Por ahora, suponemos que los datos son linealmente separables.  

El _hiperplano positivo_ se define como:  

$b +  \mathbf{w^{T}x_+} = 1$

con $\mathbf{x_+}$ Los vectores de soporte positivo.

Del mismo modo, el _hiperplano negativo_ se define como:  

$b +  \mathbf{w^{T}x_-} = -1$

El primero menos el segundo, tenemos:

$\mathbf{w^{T}(x_+ - x_-)} = 2$

Podemos normalizar por la longitud del vector $w$, definido como

$||w|| = \sqrt{\sum_{j=1}^{m} w_{j}^{2}}$

Obteniendo 

$\frac{\mathbf{w^{T}(x_+ - x_-)}}{||w||} = \frac{2}{||w||}$

El lado izquierdo se puede interpretar como la distancia entre el hiperplano positivo y negativo, que es el _margen_ que queremos maximizar.

Por lo tanto, queremos maximizar $\frac{2}{||w||}$ bajo la restricción de que todas las muestras se clasifican correctamente:

$b +  \mathbf{w^{T}x^{(i)}} \geq 1 \,\,\, if \,\,\, y^{(i)} = 1$  
$b +  \mathbf{w^{T}x^{(i)}} \leq -1 \,\,\, if \,\,\, y^{(i)} = -1$  

i.e. Todos los ejemplos negativos deben caer a un lado del hiperplano negativo y viceversa. O:

$y^{(i)} (b +  \mathbf{w^{T}x^{(i)}}) \geq 1 \,\,\, \forall{i}$  

Maximizando $\frac{2}{||w||}$ se puede hacer minimizando $\frac{||w||^2}{2}$

Este es un objetivo cuadrático con restricciones lineales y, por lo tanto, puede resolverse mediante programación cuadrática,y más específicamente con el _método de multiplicadores de Lagrange_.
* Se vuelve más complicado cuando los datos no son linealmente separables.

Interpretación geométrica
- Suponga 2 coeficientes $w_1$ y $w_2$
- Función objetivo cuadrática $f = \frac{||w||^2}{2}$
- Restricción $y^{(i)} (b +  \mathbf{w^{T}x^{(i)}}) > 1 \,\,\, \forall{i}$
    - $g(w_1, w_2) = 1$ en la imagen
- Encuentra el punto ($w_1$, $w_2$) que satisface $g$ pero maximiza $f$

![](https://drive.google.com/uc?id=1U2Jfm2oaJRG2sygrNakaIE4dGTZ7Ljv2)

##### Formulaciones primarias y duales

La formulación primaria de la función objetivo lagrangiana es minimizar:

$$\mathcal{L}_{Primal} = \frac{1}{2} ||\mathbf{w}||^2 - \sum_{i=1}^{n} a_i y_i (\mathbf{x_i} * \mathbf{w} + b) + \sum_{i=1}^{n} a_i $$

tal que

$$ a_i \geq 0 $$ 
$$ \mathbf{w} = \sum_{i=1}^{n} a_i y_i \mathbf{x_i} $$
$$ \sum_{i=1}^{n} a_i y_i = 0 $$

con $n$ el número de ejemplos de entrenamiento y $a$ las *variables duales*, que actúan como pesos para cada ejemplo de entrenamiento. Primero encontramos el conjunto óptimo de $a$, luego los $w$ se pueden calcular fácilmente.

Por lo tanto, en lugar de aprender los pesos, aprendemos qué
ejemplos son importantes y el modelo sigue a partir de eso.

Tiene una formulación dual para minimizar (Ver 'Elements of Statistical Learning'):

$$\mathcal{L}_{Dual} (a_i) = \sum_{i=1}^{l} a_i - \frac{1}{2} \sum_{i,j=1}^{l} a_i a_j y_i y_j (\mathbf{x_i} . \mathbf{x_j}) $$

Tal que

$$ a_i \geq 0 $$  
$$ \sum_{i=1}^{l} a_i y_i = 0 $$

* Observar el término $\mathbf{x_i} . \mathbf{x_j}$. Cuando las entradas solo aparecen en productos internos, llamamos al método _kernelized_ (dual kernelizada)

Estos son 2 problemas de optimización muy diferentes:
- En el primario, optimizamos $p$ variables (número de características)
- En el dual, optimizamos las $n$ variables (número de instancias)

¿Por qué estamos considerando esto?

- En algunos problemas, tenemos más funciones que puntos de datos.
- Podemos resolver el problema simplemente computando los productos internos de $\mathbf{x_i} . \mathbf{x_j}$, lo cual será importante cuando queramos resolver casos separables no lineales.

En `sklearn`, el `LinearSVC` te permite elegir entre el enfoque primario y el dual, mientras que `SVC` siempre utiliza el dual

### Haciendo predicciones

- $a_i$ será *0* si el punto se encuentra fuera del margen, en el lado derecho del límite (o frontera) de decisión
- Las muestras de entrenamiento para las cuales $a_i$ no es 0 son los _vectores de soporte_
- Por lo tanto, el modelo SVM está completamente definido por los vectores de soporte y sus coeficientes.

- Conociendo los coeficientes duales $a_i$ (de los cuales $l$ no son cero), podemos encontrar los pesos $ w $ para el hiperplano separador de máximo margen:
$$ \mathbf{w} = \sum_{i=1}^{l} a_i y_i \mathbf{x_i} $$
- Por lo tanto, podemos clasificar una nueva muestra $\mathbf{u}$ mirando el signo de $\mathbf{w}*\mathbf{u}+b$


### SVMs y kNN
Recuerde, clasificaremos una nueva muestra *u* mirando el signo de:  
$f(x) = \mathbf{w}*\mathbf{u}+b$ = $\sum_{i=1}^{l} a_i y_i \mathbf{x_i}*\mathbf{u}+b$

_Weighted k-nearest neighbor_ es una generalización del clasificador k-nearest neighbor. Clasifica puntos mirando el signo de:  
$f(x) = \sum_{i=1}^{k} a_i y_i dist(x_i, u)$

Por lo tanto: los SVM predicen exactamente de la misma manera que k-NN, solo que:
- Solo consideran los puntos verdaderamente importantes (los vectores de soporte)
    - Por lo tanto, _mucho_ más rápido
- El número de vecinos es el número de vectores de soporte.
- La función de distancia es _un producto interno de las entradas_

### SVMs en scikit-learn

- Podemos usar el clasificador `svm.SVC`
    - o `svm.SVR` para regresión
- Para construir un SVM lineal usuar `kernel=linear`

  - `dual_coef_`: los coeficientes duales $a$, i.e. los `pesos` de los vectores de soporte


### Manejo de datos no linealmente separables

- Si los datos no son linealmente separables, la maximización del margen (duro) no tiene sentido
    - Las restricciones contradecirían
- Podemos permitir infracciones de la restricción de margen mediante la introducción de una variable de _holgura_ $\xi^{(i)}$ para cada punto de dato

$b +  \mathbf{w^{T}x^{(i)}} \geq 1 - \xi^{(i)}  \,\,\, if \,\,\, y^{(i)} = 1$  
$b +  \mathbf{w^{T}x^{(i)}} \leq -1 + \xi^{(i)} \,\,\, if \,\,\, y^{(i)} = -1$  

El nuevo objetivo (a minimizar) se convierte en:

$\frac{||w||^2}{2} + C (\sum_i \xi^{(i)})$

- $C$ es una penalización por clasificación errónea
    - C Grande: grandes penalizaciones por error
    - C Pequeño: menos estricto sobre las infracciones (más regularización)

- Esto se conoce como _soft margin_ SVM (o _large margin_ SVM)
    - Algunos vectores de soporte están exactamente en el hiperplano de margen, con margen = 1
    - Otros son infractores de margen, con margen <1 y una variable de holgura positiva: $\xi^{(i)} > 0$
        - Si $\xi^{(i)} \geq 1$, estan mal clasificados

### C y regularización

-Por lo tanto, podemos usar C para controlar el tamaño del margen y ajustar el equilibrio de sesgo-varianza
    - C Pequeño: Aumenta el sesgo, reduce la varianza, más underfitting
    - C Grande: reduce el sesgo, aumenta la varianza, más sobreajuste (overfitting)
- El término penalizador $C (\sum_i \xi^{(i)})$ actúa como un regularizador L1 en los coeficientes duales
    - También conocido como _hinge loss_
    - Esto induce la *sparsity* los valores de C grandes establecerán muchos coeficientes duales en 0, por lo tanto, menos vectores de soporte
    - Los valores pequeños de C generalmente conducirán a más vectores de soporte (más puntos caen dentro del margen)
    - De nuevo, depende de los datos qué tan flexible o estricto se debe ser
-  El _least squares SVM_ es una variante que hace la regularización L2
    - Tendrá muchos más vectores de soporte (con pesos bajos)
    - En scikit-learn, esto solo está disponible para el clasificador `LinearSVC` (`loss='squared_hinge'`)

# Kernelización
* Un método es _kernelizado_ si las entradas solo aparecen dentro de los productos internos $x_i \cdot x_j$ = $\langle x_i , x_j \rangle$
* Dijimos que los SVMs pueden ser _kernelizados_ a través de su formulación dual:

$$\mathcal{L}_{Dual} (a_i) = \sum_{i=1}^{l} a_i - \frac{1}{2} \sum_{i,j=1}^{l} a_i a_j y_i y_j (\mathbf{x_i} . \mathbf{x_j}) $$

* Una _función kernel_ correspondiente a una transformación $\Phi$ es

$$ k(x_i,x_j) = \langle \Phi(x_i) , \Phi(x_j) \rangle $$

* $\Phi(x)$ se puede utilizar para generar muchas más características en base a la característica original $x$
* Resulta que a menudo podemos evaluar $k(x_i,x_j)$ directamente, _sin evaluar_ $\Phi(x_i) , \Phi(x_j)$

## Kernel trick
* Evaluar el kernel directamente puede ser _mucho más_ barato.
* Por ejemplo: un simple mapa de atributos _cuadrático_ para $x = (x1,..., xd )$ tiene dimensión $\mathcal{O}(d^2)$:

$$ \Phi(x) = (x_1,..., x_d , x_1^2,..., x_d^2 , \sqrt{2} x_1 x_2 , ..., \sqrt{2} x_{d-1} x_{d}) $$

* El kernel cuadrático correspondiente es:

$$ k(x_i,x_j) = \langle \Phi(x_i) , \Phi(x_j) \rangle = \langle x_i , x_j \rangle + \langle x_i , x_j \rangle^2$$

* Nos podemos saltar la computación de $\Phi(x_i)$ y $\Phi(x_j)$ y computar $k(x_i,x_j)$ en $\mathcal{O}(d)$ en vez de  $\mathcal{O}(d^2)$!

## Funciones kernel
* Es útil pensar en un kernel como un puntaje de similitud entre 2 vectores (puntos)
    * No es matemáticamente equivalente
* Hay muchas formas de diseñar una puntuación de similitud (también para texto, gráficos, ...)
* Computacionalmente :_mucho_ más barato
* Podemos acceder a espacios de características muy grandes (incluso infinitos) $\mathcal{H}$
* Pensar en términos de similitud es mucho más intuitivo que pensar en espacios de características de alta dimensión

### Kernel lineales
* El espacio de entrada es igual al espacio de salida: $X = \mathcal{H} = \mathbb{R}^d$
* Mapa de características $\Phi(x) = x$
* Kernel: $ k(x_i,x_j) = x_i \cdot x_j = x_i^T x_j $
* Geométricamente, podemos verlos como _proyecciones_ de $x_j$ en un hiperplano definido por $x_i$
    * Los puntos cercanos tendrán proyecciones cercanas

## Matriz kernel
Para puntos de $x_1,..., x_n \in X$ y un producto interno $\langle \cdot , \cdot \rangle$, la matriz kernel (o _Gram matrix_) se define como:

$$ 
K =  \begin{bmatrix}
                    \langle x_1 , x_1 \rangle & \ldots & \langle x_1 , x_n \rangle \\
                    \vdots & \ddots & \vdots \\
                    \langle x_n , x_1 \rangle & \ldots & \langle x_n , x_n \rangle
                  \end{bmatrix} \\
                  $$
                  
* Para el producto interno euclidiano $ k(x_i,x_j) = x_i^T x_j $, tenemos $K=XX^T$
* Tamaño es $n x n$, independientemente del número de dimensiones $d$
* Una vez que se calcula la matriz kernel, el costo depende solo de la cantidad de puntos de datos


### Kernels

Más generalmente, un kernel en un espacio X es una función 

$k: X \times X \rightarrow \mathbb{R}$

Con las propiedades: 

- Simetría: $k(x_1,x_2) = k(x_2,x_1) \,\,\, \forall x_1,x_2 \in X$
- Definida positiva: para cada subconjunto finito de puntos de datos ${x_1,...,x_n}$, el kernel Gram matrix es semidefinida positiva
    - Intuitivamente, $k(x_1,x_2) \geq 0$

Esto es también lo que esperamos de las funciones de similitud.

#### Kernels: ejemplos

- El producto interno es un kernel. El producto interno estándar es el __linear kernel__:  
$k(x_1,x_2) = x_1^T x_2$

- Kernels pueden construirse a partir de otros kernels $k_1$ y $k_2$:
    - Para $\lambda \geq 0, \lambda . k_1$ es un kernel
    - $k_1 + k_2$ es un kernel
    - $k_1 . k_2$ es un kernel (por lo tanto también $k_1^n$)
- Esto permite construir el __polynomial kernel__:  
$k(x_1,x_2) = (x_1^T x_2 + b)^d$, for $b \geq 0$ and $d \in \mathbb{N}$ 

- El kernel 'radial base function' (o __Gaussian__) se define como:  
$k(x_1,x_2) = exp(-\gamma ||x_1 - x_2||^2)$, for $\gamma \geq 0$ 

gamma

![](https://drive.google.com/uc?id=1hVWUNcncAH1FNisOecPmaQ3j22o1pF2K)

#### El Kernel Trick (resumen)

* Agregar explícitamente características no lineales puede hacer que los modelos lineales sean mucho más potentes, pero también mucho más caros
* Dado un algoritmo ML kernelizado, Podemos cambiar el producto interno por una nueva función kernel.
* Funciones kernel nos permiten calcular directamente distancias (productos escalares) en un espacio implícito de alta dimensión (en realidad no se construye)
* Una _función kernel_ es una función de distancia (similitud) con propiedades especiales para las cuales es posible este truco
    * Kernel polinomial: calcula todos los polinomios hasta cierto grado de las características originales
    * Kernel Gaussiano, o _radial basis function (RBF)_: considera todos los polinomios posibles de todos los grados
        * Espacio infinito de alta dimensión (espacio de Hilbert), donde la importancia de las características disminuye para grados más altos

### Kernel Gaussiano: intuición
- Cada punto genera una función, el producto interno es donde se cruzan
- Cuanto más cerca están los puntos, más similares son

![](https://drive.google.com/uc?id=1LEQ--5iwpbJNAnhwIPOR8SQtsbD6tsQX)

Ejemplo (para regresión):
* Tenemos 6 puntos de entrada: [-6,-4,-3,0,2,4]
    * Ajustamos un kernel sobre cada uno (azul)
* Aprendemos un coeficiente para cada uno: ej. [-.8,.5,-0.5,.7,0.3,-0.2]
* Prediciones resultantes (curva roja)

````python
import scipy.stats as stats
import math

xs = [-6,-4,-3,0,2,4]
alphas = [-.8,.5,-0.5,.7,0.3,-0.2]
variance = 1
sigma = math.sqrt(variance)
p = np.linspace(-10, 10, 100)
f = [0]*100
for a,x in zip(alphas,xs):
    k = stats.norm.pdf(p, x, sigma)
    f += a*k
    plt.plot(p, k, 'b-')
plt.plot(p, f, 'r-')
plt.plot()
plt.show()
````

### Kernels locales vs globales
* Con un kernel lineal o polinomial, un vector de soporte puede afectar todo el espacio de modelo
    - Estos se denominan _kernels globales_
* El kernel RBF solo afecta la región alrededor del vector de soporte (dependiendo de qué tan ancho sea)
    - Esto se llama kernel _local_
    - Puede capturar anomalías locales que un kernel global no puede
    - También se sobreajusta fácilmente si los kernels son muy estrechos 

### Parámetros del Kernel

Para hacer una predicción para un nuevo punto, se mide la distancia a cada uno de los vectores de soporte.

* El peso de cada vector de soporte se almacena en el atributo `dual_coef_` de SVC
* La distancia entre los puntos de datos se mide mediante el kernel
  * Kernel Gaussiano : $krbf(x_1, x_2) = \exp(\gamma ||x_1 - x_2||^2)$
      * $\gamma$ controla el ancho del kernel y se puede ajustar

Dados los vectores de soporte, sus pesos y el kernel, podemos graficar el límite de decisión

````python
from sklearn.svm import SVC

X, y = mglearn.tools.make_handcrafted_dataset()                                                                  
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)                                                
mglearn.plots.plot_2d_separator(svm, X, eps=.5)
# plot data
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
# plot support vectors
sv = svm.support_vectors_
# class labels of support vectors are given by the sign of the dual coefficients
sv_labels = svm.dual_coef_.ravel() > 0
mglearn.discrete_scatter(sv[:, 0], sv[:, 1], sv_labels, s=15, markeredgewidth=3)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1");
````

### Afinando (tuning) de los parámetros del  SVM
Varios parámetros importantes:

* gamma ((inverso) ancho del Kernel): valores altos significa que los puntos están más separados
    * Los valores altos significan Gaussianas estrechas, es decir, la influencia de un punto es muy pequeña
        * Necesitas muchos vectores de soporte
    * Conduce a límites de decisión complejos, sobreajuste
* C (nuestro regularizador lineal): 'costo' de clasificar erróneamente ejemplos de entrenamiento
    * C Grande: fuerza a SVM para clasificar correctamente más ejemplos 
        * Requiere más vectores de soporte, por lo tanto, límites de decisión complejos
* Para kernels polinomiales, el _degree_ (exponente) define la complejidad de los modelos

````python
plt.rcParams.update({'font.size': 14})
fig, axes = plt.subplots(3, 3, figsize=(15, 10))

for ax, C in zip(axes, [-1, 0, 3]):
    for a, gamma in zip(ax, range(-1, 2)):
        mglearn.plots.plot_svm(log_C=C, log_gamma=gamma, ax=a)
        
axes[0, 0].legend(["class 0", "class 1", "sv class 0", "sv class 1"],
                  ncol=4, loc=(.9, 1.2));
````

* Gamma bajo (izquierda): Gaussianas anchas, límites de decisión muy suaves
* Gamma alto (derecha): Gaussianas estrechas, los límites se centran en puntos únicos (alta complejidad)
* C bajo (top): cada vector de soporte tiene una influencia muy limitada: muchos vectores de soporte, límite de decisión casi lineal
* C alto (bottom): Mayor influencia, el límite de decisión se dobla a cada vector de soporte

````python
#bloque para generar los plots de los clasificadores
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles, make_classification
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

def plot_classifiers(names, classifiers, figuresize):
    h = .02  # step size in the mesh
    X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
    rng = np.random.RandomState(2)
    X += 2 * rng.uniform(size=X.shape)
    linearly_separable = (X, y)

    datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable
            ]
    figure = plt.figure(figsize=figuresize)
    i = 1
    
    # iterate over datasets
    for ds_cnt, ds in enumerate(datasets):
        # preprocess dataset, split into training and test part
        X, y = ds
        X = StandardScaler().fit_transform(X)
        X_train, X_test, y_train, y_test = \
            train_test_split(X, y, test_size=.4, random_state=42)

        x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
        y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
        xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                             np.arange(y_min, y_max, h))

        # just plot the dataset first
        cm = plt.cm.RdBu
        cm_bright = ListedColormap(['#FF0000', '#0000FF'])
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        if ds_cnt == 0:
            ax.set_title("Input data")
        # Plot the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6)
        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        i += 1
    
        # iterate over classifiers
        for name, clf in zip(names, classifiers):
            ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
            clf.fit(X_train, y_train)
            score = clf.score(X_test, y_test)

            # Plot the decision boundary. For that, we will assign a color to each
            # point in the mesh [x_min, x_max]x[y_min, y_max].
            if hasattr(clf, "decision_function"):
                Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
            else:
                Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

            # Put the result into a color plot
            Z = Z.reshape(xx.shape)
            ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

            # Plot also the training points
            ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
            # and testing points
            ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                       alpha=0.6)

            ax.set_xlim(xx.min(), xx.max())
            ax.set_ylim(yy.min(), yy.max())
            ax.set_xticks(())
            ax.set_yticks(())
            if ds_cnt == 0:
                ax.set_title(name)
            ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                    size=15, horizontalalignment='right')
            i += 1
        
    plt.tight_layout()
    plt.show()
````

````python
names = ["Linear SVM", "RBF", "Polynomial"]

classifiers = [
    SVC(kernel="linear", C=0.025),
    SVC(gamma=2, C=1),
    SVC(kernel="poly", degree=3, C=0.1)
    ]
 
plot_classifiers(names, classifiers, figuresize=(20,8)) 
````

### Preprocesamiento de datos para SVM
* Los SVM son muy sensibles a la configuración de los hiperparámetros
* Esperan que todas las características estén aproximadamente en la misma escala
* Punto de similitud de datos (ej. RBF kernel) se calcula de la misma manera en todas las dimensiones
* Si alguna dimensión se escala de manera diferente, tendrá un impacto mucho mayor / menor

````python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=0)

svc = SVC()
svc.fit(X_train, y_train)

print("Accuracy on training set: {:.2f}".format(svc.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(svc.score(X_test, y_test))) 
````

    * Aplíquelos en la partición entrenamiento y pruebe por separado
* sklearn ofrece `pipelines` que hacen esto más fácil
    * Wrapper alrededor de una serie de operadores

### Fortalezas, debilidades y parámetros

* Los SVM permiten límites de decisión complejos, incluso con pocas características.
* Funciona bien en datos de baja y alta dimensión.

* No escala muy bien a grandes conjuntos de datos (> 100000)
* Requiere un cuidadoso preprocesamiento de los datos y el ajuste de los parámetros.
* Los modelos SVM son difíciles de inspeccionar

Parámetros importantes:
* parámetro de regularización _C_
* elección del kernel y parámetros específicos del kernel
    * Ttípicamente fuerte correlación con _C_

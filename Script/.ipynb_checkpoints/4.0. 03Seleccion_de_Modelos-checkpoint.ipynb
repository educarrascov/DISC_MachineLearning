## The holdout (partición simple train-test)
Ya hemos visto la forma más básica de evaluación:

- _Aleatoriamente_ divide los datos (y las etiquetas correspondientes) en conjunto de entrenamiento y prueba (75% -25%)
- Entrenar (fit) un modelo con los datos de entrenamiento
- Evalúe (score) un modelo en los datos de prueba (comparando etiquetas pronosticadas y verdaderas)
    - Estamos interesados en qué tan bien el modelo _generaliza_ a los datos nuevos (de test)

En scikit-learn: `train_test_split`

``` python
X, y = make_blobs(centers=2, random_state=0)
# split data and labels into a training and a test set

# Fit a model to the training set
model = LogisticRegression().fit(X_train, y_train)
# Evaluate on the test data
test_score = model.score(X_test, y_test)
```

````python
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# create a synthetic dataset
X, y = make_blobs(centers=2, random_state=0)
# split data and labels into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
# Instantiate a model and fit it to the training set
model = LogisticRegression().fit(X_train, y_train)
# evaluate the model on the test set
print("Test set score: {:.2f}".format(model.score(X_test, y_test)))
````

##### Cross-validation
- ¿Qué pasa si una división aleatoria produce diferentes modelos (y scores) que otra?
- Reduzca el sesgo probando en cada punto exactamente una vez
- _Validación cruzada de k-dobladas_ o _k-fold cross-validation_ (CV): particionar (split) aleatoriamente en _k_ partes del mismo tamaño, llamadas dobladas _folds_. Supongamos _k_=5:
    - Primero, el fold 1 es el conjunto de prueba, y los folds 2-5 comprenden el conjunto de entrenamiento
    - Luego, el fold 2 es el conjunto de prueba, los folds 1,3,4,5 comprenden el conjunto de entrenamiento
    - Calcule las _k_ puntuaciones de evaluación, agregue después (por ejemplo, tome la media)

````python
mglearn.plots.plot_cross_validation()
````

En scikit-learn:
- La función `cross_val_score` con el algortimo de aprendizaje, datos, etiquetas, número de folds
- Devuelve la lista de todos los scores. Los modelos se construyen internamente, pero no se devuelven
- Valores por defecto: 3-fold CV, accuracy (clasificación) o $R^2$ (regresión)
- Tener en cuenta que puede haber bastante _varianza_ en los resultados
    - Depende de la estabilidad del modelo y la cantidad de datos de entrenamiento
    - Por lo general, cuanto más datos de entrenamiento, más estables son los modelos

``` python
logreg = LogisticRegression()
scores = cross_val_score(logreg, iris.data, iris.target, cv=5)
```

````python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()
logreg = LogisticRegression()

scores = cross_val_score(logreg, iris.data, iris.target, cv=5)
print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))
print("Variance in cross-validation score: {:.4f}".format(np.var(scores)))
````

### Beneficios y desventajas de cross-validation
- Más robusto: cada ejemplo de entrenamiento estará en un conjunto de prueba exactamente una vez
    - El modelo se evalúa en todas las muestras, debe funcionar bien en todas
    - En una partición train-test, podemos tener
        - 'suerte': todos los ejemplos sencillos están en el conjunto de prueba
        - 'desafortunado': todos los ejemplos difíciles están en el conjunto de prueba
- Muestra cuán _sensible_ es el modelo al conjunto de entrenamiento exacto
- Mejor estimación del rendimiento real
    - 10-fold CV utiliza el 90% de todos los datos para el entrenamiento (frente al 75% con holdout)
    - Mientras más alto _k_, más precisa es la estimación
- Desventaja: costo computacional, aproximadamente _k_ veces más lento que el holdout
- Los modelos inestables (por ejemplo, deep learning) pueden no converger en cada fold

### Stratified K-Fold cross-validation

- Si los datos están *desequilibrados*, algunas clases tienen muchas menos muestras
- Probable que algunas clases no estén presentes en el conjunto de pruebas 
- Estratificación: _proporciones_ entre clases se conservan en cada fold
    - Ordenar ejemplos por clase
    - Separe las muestras de cada clase en _k_ conjuntos (estratos)
    - Combina el estrato correspondiente en folds

````python
mglearn.plots.plot_stratified_cross_validation()
````

En scikit-learn:
- Utiliza validación cruzada estratificada por defecto para la clasificación
- Validación cruzada normal para regresión
- Ambos no son aleatorios (las muestras no se mezclan de antemano)
    - Los datos ordenados (por ejemplo, series de tiempo) nunca deben ser aleatorizados
- Puedes construir folds manualmente con `KFold` o `StratifiedKFold`
    - aleatorizable (parámetro `shuffle`)
    
``` python
kfold = KFold(n_splits=5, shuffle=False) # Not stratified
cross_val_score(logreg, iris.data, iris.target, cv=kfold)
skfold = StratifiedKFold(n_splits=5, shuffle=True)
cross_val_score(logreg, iris.data, iris.target, cv=skfold)
```

````python
from sklearn.model_selection import KFold, StratifiedKFold
kfold = KFold(n_splits=5)
print("Cross-validation scores KFold(n_splits=5):\n{}".format(
      cross_val_score(logreg, iris.data, iris.target, cv=kfold)))
skfold = StratifiedKFold(n_splits=5, shuffle=True)
print("Cross-validation scores StratifiedKFold(n_splits=5, shuffle=True):\n{}".format(
      cross_val_score(logreg, iris.data, iris.target, cv=skfold)))
````

### Leave-One-Out cross-validation

- _k_ fold cross-validation con _k_ igual al número de muestras
- Completamente imparcial (en términos de divisiones de datos), pero computacionalmente costoso
- Pero: generaliza _menos_ bien hacia los datos no visto
    - Los conjuntos de entrenamiento están correlacionados (se superponen en gran medida)
    - Sobreajusta en los datos utilizados para la (totalidad) evaluación
    - Una muestra diferente de los datos puede arrojar resultados diferentes.
- Recomendado solo para conjuntos de datos pequeños

``` python
loo = LeaveOneOut()
scores = cross_val_score(logreg, iris.data, iris.target, cv=loo)
```

````python
from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
scores = cross_val_score(logreg, iris.data, iris.target, cv=loo)
print("Number of cv iterations: ", len(scores))
print("Mean accuracy: {:.2f}".format(scores.mean()))
````

### Shuffle-Split cross-validation

- Muestrea un número de muestras (`train_size`) al azar como conjunto de entrenamiento, y un número disjunto de muestras (` test_size`) como conjunto de prueba
- Repita este procedimiento `n_iter` veces, obteniendo ` n_iter` scores
- Práctico cuando se usan conjuntos de datos muy grandes

- Ejemplo con `train_size` = 5, `test_size` = 2, `n_iter` = 4

shuffle_split

shuffle_split = StratifiedShuffleSplit(test_size=.5, train_size=.5, n_splits=10)

Nota: esto está relacionado a *bootstrapping*:
- Muestrea _n_ (número total de muestras) puntos de datos, con reemplazo, como conjunto de entrenamiento (el bootstrap)
- Utilice las muestras no muestreadas (out-of-bootstrap) como conjunto de prueba
- Repetir `n_iter` veces, obteniendo `n_iter` puntajes (scores)
- No disponible directamente en scikit-learn, en vez utilizar Shuffle-Split.
    - Con `train_size=0.66`, `test_size=0.34`
    - Se puede demostrar que bootstraps incluyen el 66% de todos los puntos en promedio 

### Repeated cross-validation
- Cross-validation todavía está sesgado porque la división inicial se puede hacer de muchas maneras
- Repetido, o n-times-k-fold cross-validation:
    - Mezcle datos aleatoriamente, luejo hacer k-fold cross-validation
    - Repetir n veces, produce n veces k puntajes
- Imparcial, muy robusto, pero n veces más caro

### La elección de un procedimiento de estimación de rendimiento
Sin reglas estrictas, solo algunas guías:

- Utilice siempre la estratificación para la clasificación
- Utilizar holdout para conjuntos de datos muy grandes (ej. >1.000.000 ejemplos)
    - O cuando el algoritmo de ML no siempre converge(ej. deep learning)
- Elegir _k_ dependiendo del tamaño del conjunto de datos y los recursos
    - Utilizar leave-one-out para conjuntos de datos pequeños (ej. <500 ejemplos)
    - Utilizar cross-validation para los otros casos
        - Más popular (y teóricamente sólido): 10-fold CV
        - La literatura sugiere 5x2-fold CV es mejor

### Evaluación vs Optimización

- Cada algoritmo optimiza una función objetivo dada (en los datos de entrenamiento)
    - Ej. recordar la pérdida L2 en la regresión de Ridge
$$\mathcal{L}_{ridge} = \sum_{i}(y_i - \sum_{j} x_{i,j}w_j-b)^2 + \alpha \sum_{i} w_i^2$$
- La elección de la función está limitada por lo que se puede optimizar de manera eficiente
    - Ej. gradiente descendente requiere una función de pérdida diferenciable
- _Evaluamos_ el modelo resultante con una puntuación (score) que tiene sentido en el mundo real
    - Ej. porcentaje de predicciones correctas (en un conjunto de prueba)
- También ajustamos los hiperparámetros del algoritmo para maximizar esa puntuación

### Clasificación binaria

- Se tiene una clase positiva y una negativa
- 2 tipos diferentes de errores:
    - Falso Positivo (Error tipo I): el modelo predice positivo mientras que la etiqueta verdadera es negativa
    - Falso Negativo (Error tipo II): el modelo predice negativo mientras que la etiqueta verdadera es positiva
- No siempre son igualmente importantes
    - ¿De qué lado quieres equivocarte para un examen médico?

#### Matrices de confusión

- Podemos representar todas las predicciones (correctas e incorrectas) en una matriz de confusión
    - una matriz de n por n (n es el número de clases)
    - Las filas corresponden a clases verdaderas, columnas a clases predichas
    - Cada entrada cuenta con qué frecuencia una muestra que pertenece a la clase correspondiente a la fila se clasificó como la clase correspondiente a la columna.
    - Para la clasificación binaria, etiquetamos estos verdaderos negativos (TN), verdaderos positivos (TP), falsos negativos (FN), falsos positivos (FP)

````python
mglearn.plots.plot_binary_confusion_matrix()
````

#### Precisión predictiva (accuracy)

- Accuracy es una de las medidas que podemos calcular en función de la matriz de confusión:

\begin{equation}
\text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{TP} + \text{TN} + \text{FP} + \text{FN}}
\end{equation}

- En sklearn: utilice `confusion_matrix` y `accuracy_score` desde `sklearn.metrics`.
- Accuracy también es la medida de evaluación predeterminada (por default) para la clasificación

````python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
data = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, stratify=data.target, random_state=0)
lr = LogisticRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)

print("confusion_matrix(y_test, y_pred): \n", confusion_matrix(y_test, y_pred))
print("accuracy_score(y_test, y_pred): ", accuracy_score(y_test, y_pred))
print("model.score(X_test, y_test): ", lr.score(X_test, y_test))
````

#### El problema con accuracy: conjuntos de datos desequilibrados (desbalanceados)

- El tipo de error juega un papel aún mayor si el conjunto de datos está desbalanceado
    - Una clase es mucho más frecuente que la otra, por ejemplo, fraude crediticio
    - ¿Es un accuracy del 99.99% lo suficientemente buena? 
- ¿Son estos tres modelos realmente igual de buenos?

````python
# From Applied Machine Learning
def plot_confusion_matrix(values, xlabel="predicted labels", ylabel="true labels", xticklabels=None,
                          yticklabels=None, cmap=None, vmin=None, vmax=None, ax=None,
                          fmt="{:.2f}", xtickrotation=45, norm=None):
    """Plot a matrix as heatmap with explicit numbers.
    Parameters
    ----------
    values : ndarray
        Two-dimensional array to visualize.
    xlabel : string, default=""
        Label for the x-axis.
    ylabel : string, default=""
        Label for the y-axis.
    xticklabels : list of string or None, default=None
        Tick labels for the x-axis.
    yticklabels : list of string or None, default=None
        Tick labels for the y-axis
    cmap : string or colormap
        Matpotlib colormap to use.
    vmin : int, float or None
        Minimum clipping value.
    vmax : int, float or None
        Maximum clipping value.
    ax : axes object or None
        Matplotlib axes object to plot into. If None, the current axes are
        used.
    fmt : string, default="{:.2f}"
        Format string to convert value to text.
    xtickrotation : float, default=45
        Rotation of the xticklabels.
    norm : matplotlib normalizer
        Normalizer passed to pcolor
    """
    import matplotlib.pyplot as plt
    if ax is None:
        ax = plt.gca()
    img = ax.pcolormesh(values, cmap=cmap, vmin=vmin, vmax=vmax, norm=norm)
    # this will allow us to access the pixel values:
    img.update_scalarmappable()
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    ax.set_xlim(0, values.shape[1])
    ax.set_ylim(0, values.shape[0])

    if xticklabels is None:
        xticklabels = [""] * values.shape[1]
    if yticklabels is None:
        yticklabels = [""] * values.shape[0]

    # +.5 makes the ticks centered on the pixels
    ax.set_xticks(np.arange(values.shape[1]) + .5)
    ax.set_xticklabels(xticklabels, ha="center", rotation=xtickrotation)
    ax.set_yticks(np.arange(values.shape[0]) + .5)
    ax.set_yticklabels(yticklabels, va="center")
    ax.set_aspect(1)

    for p, color, value in zip(img.get_paths(), img.get_facecolors(),
                               img.get_array()):
        x, y = p.vertices[:-2, :].mean(0)
        if np.mean(color[:3]) > 0.5:
            # pixel bright: use black for number
            c = 'k'
        else:
            c = 'w'
        ax.text(x, y, fmt.format(value), color=c, ha="center", va="center")
    ax.invert_yaxis()
    return ax
````

````python
# Artificial 90-10 imbalanced target
y_true = np.zeros(100, dtype=int)
y_true[:10] = 1
y_pred_1 = np.zeros(100, dtype=int)
y_pred_2 = y_true.copy()
y_pred_2[10:20] = 1
y_pred_3 = y_true.copy()
y_pred_3[5:15] = 1 - y_pred_3[5:15]

def plot_measure(measure):
    fig, axes = plt.subplots(1, 3)
    for i, (ax, y_pred) in enumerate(zip(axes, [y_pred_1, y_pred_2, y_pred_3])):
        plot_confusion_matrix(confusion_matrix(y_true, y_pred), cmap='gray_r', ax=ax,
                              xticklabels=["N", "P"], yticklabels=["N", "P"], xtickrotation=0, vmin=0, vmax=100)
        ax.set_title("{}: {:.2f}".format(measure.__name__,measure(y_true, y_pred)))

    plt.tight_layout()
    
plt.rcParams['figure.dpi'] = 120 # Use 300 for PDF, 100 for slides
````

````python
plot_measure(accuracy_score)
````

__Precision__ se usa cuando el objetivo es limitar los FPs
- Ensayos clínicos: solo desea probar medicamentos que realmente funcionan
- Motores de búsqueda: desea evitar malos resultados de búsqueda

\begin{equation}
\text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
\end{equation}

````python
from sklearn.metrics import precision_score
plot_measure(precision_score)
````

__Recall__ se usa cuando el objetivo es limitar los FNs
- Diagnóstico de cáncer: uno no quiere perderse una enfermedad grave
- Motores de búsqueda: no quieres omitir resultados importantes
- También conocido como sensibilidad (sensitivity), hit rate, true positive rate (TPR)

\begin{equation}
\text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
\end{equation}

````python
from sklearn.metrics import recall_score
plot_measure(recall_score)
````

Promedio de puntajes por clase
- Estudiar los puntajes _por clase_ (en scikit-learn: `classification_report`)
    - Una clase vista como positiva, otra (s) como negativa
    - Soporte (Support): número de muestras en cada clase

- Promedio para la medida de puntuación R en las clases C (también para multiclase):
  - micro: contar el número total de TP, FP, TN, FN
  - macro $$\frac{1}{C} \sum_{c \in C} R(y_c,\hat{y_c})$$

## Teniendo en cuenta la incertidumbre

- Recuerde que muchos clasificadores en realidad devuelven una probabilidad por clase
    - La podemos recuperar con `decision_function` y `predict_proba`
- Para la clasificación binaria, usamos un umbral de 0 para  `decision_function` y 0.5 para `predict_proba` por defecto
- Sin embargo, dependiendo de la medida de evaluación, es posible que se desee establecer un umbral diferente para ajustarse a las metas propuestas
    - Por ejemplo, cuando un FP es mucho peor que un FN
    - Esto se llama _calibración del umbral_

- Imaginemos que queremos evitar clasificar erróneamente un punto positivo (rojo)
- Los puntos dentro de la frontera de decisión (línea negra) se clasifican como positivos
- Reducir el umbral de decisión (figura inferior): menos FN, más FP

### Curvas de Precision-Recall

- El mejor umbral depende de la aplicación, debe ser impulsado por objetivos del mundo real.
- Puede tener una alto recall arbitrariamente, pero a menudo también se desea una precisión razonable.
- No está claro de antemano dónde estará la compensación óptima (o el *punto de operación*), por lo que es útil observar todos los umbrales posibles
- Graficando precision versus recall para todos los umbrales nos da una __curva de precision-recall__

- En scikit-learn, esto está incluido en el módulo `sklearn.metrics`
- Retorna todos los valores de precision y recall para todos los umbrales
    - Varíe el umbral del puntaje de la función de decisión más bajo al más alto en las predicciones
    - O de mayor a menor probabilidad de clase

````python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(
    y_test, svc.decision_function(X_test)) 
````

- La compensación predeterminada (elegida por el método `predict`) se muestra como _threshold zero_.
    - Umbral más alto, más precision (mover hacia la izquierda)
    - Umbral más bajo, más recall (mover hacia la derecha)
- Cuanto más cerca esté la curva de la esquina superior derecha, mejor
    - Alta precision y alto recall
- Aquí, es posible obtener una precision de 0.5 con un alto recall

````python
# create a similar dataset as before, but with more samples
# to get a smoother curve
X, y = make_blobs(n_samples=(4000, 500), centers=2, cluster_std=[7.0, 2],
                  random_state=22)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

svc = SVC(gamma=.05).fit(X_train, y_train)

precision, recall, thresholds = precision_recall_curve(
    y_test, svc.decision_function(X_test))
# find threshold closest to zero:
close_zero = np.argmin(np.abs(thresholds))
plt.plot(recall[close_zero], precision[close_zero], 'o', markersize=10,
         label="threshold zero", fillstyle="none", c='k', mew=2)

plt.plot(recall, precision, label="precision recall curve")
plt.ylabel("Precision")
plt.xlabel("Recall")
plt.legend(loc="best");
````

#### Selección de modelo

- Los diferentes clasificadores funcionan mejor en diferentes partes de la curva (en diferentes puntos de operación)
- RandomForest (en rojo) funciona mejor en los extremos, SVM mejor en el centro
- El área debajo de la curva precision-recall (AUPRC) a menudo se usa como una medida de evaluación general

````python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=0, max_features=2)
rf.fit(X_train, y_train)

# RandomForestClassifier has predict_proba, but not decision_function
# Only pass probabilities for the positive class
precision_rf, recall_rf, thresholds_rf = precision_recall_curve(
    y_test, rf.predict_proba(X_test)[:, 1])

plt.plot(recall, precision, label="svc")

plt.plot(recall[close_zero], precision[close_zero], 'o', markersize=10,
         label="threshold zero svc", fillstyle="none", c='k', mew=2)

plt.plot(recall_rf, precision_rf, label="rf")

close_default_rf = np.argmin(np.abs(thresholds_rf - 0.5))
plt.plot( recall_rf[close_default_rf], precision_rf[close_default_rf], '^', c='k',
         markersize=10, label="threshold 0.5 rf", fillstyle="none", mew=2)
plt.ylabel("Precision")
plt.xlabel("Recall")
plt.legend(loc="best");
````

### Receiver Operating Characteristics (ROC) y AUC

- Hay otra compensación entre el recall (tasa de verdaderos positivos, TPR) y la tasa de falsos positivos (FPR).
- El espacio 2D creado por TPR y FPR se llama el espacio Receiver Operating Characteristics (ROC)
- Un modelo estará en un punto en este espacio ROC

\begin{equation}
\text{TPR} = \frac{\text{TP}}{\text{TP} + \text{FN}}
\end{equation}

\begin{equation}
\text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}
\end{equation}

- Al variar el umbral de decisión se obtiene la curva ROC
- Se puede calcular con la función `roc_curve`
    - Umbral más bajo, más recall / TPR, mover hacia la derecha
    - Umbral más alto, menos FPs, mover hacia la izquierda
- Caso ideal está cerca de la esquina superior izquierda: alto recall, bajo FPR
- Se debe inspeccionar la curva para encontrar la calibración preferida
    - Aquí, podemos obtener un mayor recall con un FPR ligeramente peor

````python
plt.rcParams['savefig.dpi'] = 100 # Use 300 for PDF, 100 for slides
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, svc.decision_function(X_test))

plt.plot(fpr, tpr, label="ROC Curve")
plt.xlabel("FPR")
plt.ylabel("TPR (recall)")
# find threshold closest to zero:
close_zero = np.argmin(np.abs(thresholds))
plt.plot(fpr[close_zero], tpr[close_zero], 'o', markersize=10,
         label="threshold zero", fillstyle="none", c='k', mew=2)
plt.legend(loc=4);
````

#### Visualización

- La densidad de probabilidad azul muestra la probabilidad p (x) de que el modelo predice azul si un punto de datos tiene cierta probabilidad pronosticada (predicha) x para ser azul. Lo mismo para el rojo.
- En un clasificador aleatorio, las densidades de probabilidad se superponen por completo.
- Todos los puntos con una probabilidad predicha superior al umbral se predicen como positivos, otros negativos
- A medida que aumentemos el umbral, obtendremos menos FP, más FN. Nos movemos de derecha a izquierda a lo largo de la curva ROC.

#### ROC Isometrics
- Diferentes *costos* pueden estar involucrados para FP y FN
- Esto produce diferentes _isometrías_ (líneas de igual costo) en el espacio ROC
- El umbral óptimo es el punto en la curva ROC donde el costo es mínimo
    - Si un FP y FN se ponderan por igual, las líneas de costo siguen la diagonal (línea azul)
    - Si un FP es 10 veces peor que un FN: línea rosada
    - Si un FN es 10 veces peor que un FP: línea roja

Diferentes formas de calcular el promedio

- macro-averaging: calcula puntajes no ponderados por clase:
$\frac{\sum_{i=0}^{n}score_i}{n}$
    - Usar cuando existe interés por cada clase por igual

- weighted averaging: los puntajes se ponderan por el tamaño relativo de las clases (soporte): 
$\frac{\sum_{i=0}^{n}score_i weight_i}{n}$
    - Utilizar cuando los datos están desbalanceados

- micro-averaging: calcula el número total de FP, FN, TP en todas las clases, luego calcula los puntajes usando estos conteos: 
$recall = \frac{\sum_{i=0}^{n}TP_i}{\sum_{i=0}^{n}TP_i + \sum_{i=0}^{n}FN_i}$
    - Usar cuando existe interés por cada muestra por igual

````python
print("Micro average f1 score: {:.3f}".format(f1_score(y_test, pred, average="micro")))
print("Weighted average f1 score: {:.3f}".format(f1_score(y_test, pred, average="weighted")))
print("Macro average f1 score: {:.3f}".format(f1_score(y_test, pred, average="macro")))
````

## Métricas para regresión

Las más utilizadas son
- (root) mean squared error: $\frac{\sum_{i}(y_{pred_i}-y_{actual_i})^2}{n}$
- mean absolute error: $\frac{\sum_{i}|y_{pred_i}-y_{actual_i}|}{n}$
    - Menos sensible a los valores atípicos y a los grandes errores
- R squared (r2): $1 - \frac{\sum_{i}(y_{pred_i}-y_{actual_i})^2}{\sum_{i}(y_{mean}-y_{actual_i})^2}$
    - Proporción de variación explicada por el modelo / variación total
    - Entre 0 y 1, pero _negativo_ si el modelo es peor que solo predecir la media
    - Más fácil de interpretar (más alto es mejor)

- R squared: 1 - proporción de $\sum_{i}(y_{pred_i}-y_{actual_i})^2$ (blue) y $\sum_{i}(y_{mean}-y_{actual_i})^2$ (red)

## Algunos pensamientos finales para esta parte
* Existen técnicas para corregir el desbalance de etiquetas
    - Undersample la clase mayoritaria, o oversample la classe minoritaria
    - SMOTE (Synthetic Minority Oversampling TEchnique) agrega puntos de _entrenamiento_ artificiales al interpolar los puntos existentes de la clase minoritaria 
* Clasificación sensible al costo (no en sklearn)
    - *Matriz de costos*: una matriz de confusión con costos asociados a cada tipo de error posible
    - Algunos algoritmos permiten optimizar estos costos en lugar de su función de pérdida habitual
    - Meta-costo: crea un conjunto (ensamble) de modelos al volver a etiquetar conjuntos de entrenamiento para que coincidan con una matriz de costos dada
        - Black-box: puede hacer que cualquier algoritmo sea rentable (pero más lento y menos preciso)

* Hay muchas más métricas para elegir
  - Cohen's Kappa: accuracy, teniendo en cuenta la posibilidad de predecir la clase correcta por casualidad
  - 1: predicción perfecta, 0: predicción aleatoria, negativa: peor que aleatoria
  - Con $p_0$ = accuracy, y $p_e$ = accuracy de un clasificador aleatorio:
  $$\kappa = \frac{p_o - p_e}{1 - p_e}$$
    - Accuracy balanceado: accuracy donde cada muestra se pondera de acuerdo con la prevalencia inversa de su clase verdadera
        - Idéntico al recall macro-averaged
    - Coeficiente de correlación de Matthews: Otra medida que se puede utilizar en datos desbalanceados
        - 1: predicción perfecta, 0: predicción aleatoria, -1: predicción inversa
            $$MCC = \frac{tp \times tn - fp \times fn}{\sqrt{(tp + fp)(tp + fn)(tn + fp)(tn + fn)}}$$


## Descomposición de sesgo-varianza (Bias-Variance decomposition)

- Cuando repetimos los procedimientos de evaluación varias veces, podemos distinguir dos fuentes de errores:
    - Bias o sesgo: error sistemático (independiente de la muestra de entrenamiento). El clasificador siempre obtiene ciertos puntos incorrectos
    - Varianza o Variance: error debido a la variabilidad del modelo con respecto a la muestra de entrenamiento. El clasificador predice algunos puntos con precisión en algunos conjuntos de entrenamiento, pero de manera imprecisa en otros.
- También hay un error intrínseco (ruido), pero no hay nada que podamos hacer contra eso.
- Bias o sesgo está asociado con underfitting y la varianza con el overfitting
- Compensación Bias-variance: a menudo se puede intercambiar varianza por sesgo a través de la regularización (y viceversa)
    - El desafío es encontrar el equilibrio correcto o compensación (minimizando el error total)
- Útil para comprender cómo ajustar o adaptar un algoritmo de aprendizaje

![](https://drive.google.com/uc?id=1OaTLnTdnzErIfwQ9HPDeADTT-384nSkt)

- Lamentablemente, esto todavía no es compatible con scikit-learn
- Cómo medir el sesgo y la varianza (para la regresión):
    - Tomar 100 o más bootstraps (o shuffle-splits)
    - Para cada punto de dato x:
  - $bias(x)^2 = (x_{true} - mean(x_{predicted}))^2$
        - $variance(x) = var(x_{predicted})$
    - Bias total: $\sum_{x} bias(x)^2 * w_x$, con $w_x$ la proporción de que x ocurra en el conjunto de prueba
    - Varianza total: $\sum_{x} variance(x) * w_x$

- Procedimiento general para la clasificación (binaria):
    - Tomar 100 o más bootstraps (o shuffle-splits)
    - Bias para cualquier punto x = proporción de clasificaciones erróneas
  - Si se clasifica erróneamente el 50% del tiempo: $bias(x)=0.5$
    - Varianza para cualquier punto x es $(1 - (P(class_1)^2 + P(class_2)^2))/2$
        - $P(class_i)$ es la proporción de las predicciones de la clase $i$
        - Cuando cada clase predijo la mitad del tiempo: $variance(x) = (1-(0.5^2+0.5^2))/2 = 0.25$
    - Total bias: $\sum_{x} bias(x)^2 * w_x$, con $w_x$ la proporción de que x ocurra en el conjunto de prueba
    - Total variance: $\sum_{x} variance(x) * w_x$

``` python
for i, (train_index, test_index) in enumerate(shuffle_split.split(X)):
    clf.fit(X[train_index], y[train_index])
    y_pred = clf.predict(X[test_index])

```

````python
# Data and classifier
X, y = make_blobs(centers=2, n_samples=1000, random_state=0)
clf = LogisticRegression()

# Bootstraps
n_repeat = 100
shuffle_split = ShuffleSplit(test_size=0.33, n_splits=n_repeat)

# Store sample predictions
y_all_pred = [[] for _ in range(len(y))]

# Train classifier on each bootstrap and score predictions
for i, (train_index, test_index) in enumerate(shuffle_split.split(X)):
    # Train and predict
    clf.fit(X[train_index], y[train_index])
    y_pred = clf.predict(X[test_index])

    # Store predictions
    for i,index in enumerate(test_index):
        y_all_pred[index].append(y_pred[i])
        
# Compute bias, variance, error
bias_sq = sum([ (1 - x.count(y[i])/len(x))**2 * len(x)/n_repeat 
            for i,x in enumerate(y_all_pred)])
var = sum([((1 - ((x.count(0)/len(x))**2 + (x.count(1)/len(x))**2))/2) * len(x)/n_repeat
           for i,x in enumerate(y_all_pred)])
error = sum([ (1 - x.count(y[i])/len(x)) * len(x)/n_repeat 
            for i,x in enumerate(y_all_pred)])

print("Bias squared: %.2f, Variance: %.2f, Total error: %.2f" % (bias_sq, var, error))
````

- Alto sesgo (bias) significa que es probable que esté underfitting

    - Use un modelo más flexible / complejo (otro algoritmo)
    - Use una técnica de reducción de sesgo (por ejemplo, boosting, que veremos más adelante)
- Alta varianza (variance)  significa que es probable que esté overfitting
    - Usar más regularización
    - Obtén más datos
    - Use un modelo más simple (otro algoritmo)
    - Utilice técnicas de reducción de varianza (ej. bagging, que veremos más adelante)

### Grid Search

- Para cada hiperparámetro, crear una lista de valores interesantes / posibles
    - Ej. Para kNN: k in [1,3,5,7,9,11,33,55,77,99]
- Evaluar todas las combinaciones posibles de valores de hiperparámetros.
    - Ej. usando cross-validation
- Seleccione los valores de hiperparámetro para obtener los mejores resultados
- Un enfoque ingenuo sería simplemente recorrer todas las combinaciones

``` python
for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # for each combination, train and evaluate an SVC
        svm = SVC(gamma=gamma, C=C);
        svm.fit(X_train, y_train);
        score = svm.score(X_test, y_test)
        # if we got a better score, store the score and parameters
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}
```

````python
# naive grid search implementation
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                    random_state=0)
print("Size of training set: {}   size of test set: {}".format(
      X_train.shape[0], X_test.shape[0]))

best_score = 0

# Note: we've seen the `C` parameter of SVMs, we'll see the `gamma` parameter later.
for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # for each combination, train an SVC
        svm = SVC(gamma=gamma, C=C);
        svm.fit(X_train, y_train);
        # evaluate the SVC on the test set 
        score = svm.score(X_test, y_test)
        # if we got a better score, store the score and parameters
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}
            
print("Best score: {:.2f}".format(best_score))
print("Best parameters: {}".format(best_parameters))
````

#### Overfitting los parámetros y el conjunto de validación
- Simplemente tomando el modelo de mejor rendimiento produce resultados optimistas
- ¡Ya hemos usado los datos de prueba para evaluar cada configuración de hiperparámetro!
- Por lo tanto, no tenemos un conjunto de pruebas independiente para evaluar estas configuraciones de hiperparámetros
    - La información se 'filtra' del conjunto de pruebas al modelo final
- Solución: reserve parte de los datos de entrenamiento para evaluar la configuración del hiperparámetro
    - Seleccione los mejores hiperparámetros en el conjunto de validación
    - Reconstruir el modelo en el conjunto de entrenamiento + validación
    - Evaluar el modelo óptimo en el conjunto de prueba.

````python
from sklearn.svm import SVC
# split data into train+validation set and test set
X_trainval, X_test, y_trainval, y_test = train_test_split(
    iris.data, iris.target, random_state=0)
# split train+validation set into training and validation set
X_train, X_valid, y_train, y_valid = train_test_split(
    X_trainval, y_trainval, random_state=1)
print("Size of training set: {}   size of validation set: {}   size of test set:"
      " {}\n".format(X_train.shape[0], X_valid.shape[0], X_test.shape[0]))

best_score = 0

for gamma in [0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # for each combination of parameters
        # train an SVC
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)
        # evaluate the SVC on the test set 
        score = svm.score(X_valid, y_valid)
        # if we got a better score, store the score and parameters
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}

# rebuild a model on the combined training and validation set,
# and evaluate it on the test set
svm = SVC(**best_parameters)
svm.fit(X_trainval, y_trainval)
test_score = svm.score(X_test, y_test)
print("Best score on validation set: {:.2f}".format(best_score))
print("Best parameters: ", best_parameters)
print("Test set score with best parameters: {:.2f}".format(test_score))
````

#### Nested cross-validation

- Hay que tener en cuenta que todavía estamos usando una sola división para crear el conjunto de prueba externo
- También podemos usar validación cruzada aquí
- Nested cross-validation:
    - Outer loop: dividir datos en conjuntos de entrenamiento y prueba
    - Inner loop: ejecutar grid search, dividiendo los datos de entrenamiento en conjuntos de entrenamiento y validación
- El resultado es solo una lista de puntajes
    - Habrá múltiples modelos optimizados y configuraciones de hiperparámetros (no devueltos)
- Para aplicar en datos futuros, necesitamos entrenar nuevamente a 'GridSearchCV' en todos los datos

``` python
scores = cross_val_score(GridSearchCV(SVC(), param_grid, cv=5),
                         iris.data, iris.target, cv=5)
```

### Random Search

- Grid Search tiene algunas debilidades:
    - La optimización de muchos hiperparámetros crea una explosión combinatorial
    - Debe predefinir una grilla, por lo tanto, puede "saltar" sobre valores óptimos
- Random Search:
    - Elige `n_iter` valores de parámetros aleatorios
    - Escala mejor, uno controla la cantidad de iteraciones
    - A menudo también funciona mejor en la práctica
        - no todos los hiperparámetros interactúan fuertemente
        - no necesitas explorar todas las combinaciones

- Corriendo random search en scikit-learn:
    - `RandomizedSearchCV` funciona como `GridSearchCV`
    - Tiene el parámetro `n_iter` para el número de iteraciones
    - Search grid puede usar distribuciones en lugar de listas fijas
    
``` python
param_grid = {'C': expon(scale=100), 
              'gamma': expon(scale=.1)}
random_search = RandomizedSearchCV(SVC(), param_distributions=param_grid,
                                   n_iter=20)
random_search.fit(X_train, y_train)
```

````python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon

param_grid = {'C': expon(scale=100), 
              'gamma': expon(scale=.1)}
random_search = RandomizedSearchCV(SVC(), param_distributions=param_grid,
                                   n_iter=20)
X_train, X_test, y_train, y_test = train_test_split(
        iris.data, iris.target, random_state=0)
random_search.fit(X_train, y_train)
````

# Resumen

- k-fold Cross-validation
    - Elija k dependiendo de la cantidad de datos que tenga
        - k grande es más lento, pero permite más datos de entrenamiento
        - 10-fold, 5-fold, 5x2-fold son las opciones más populares
    - Utilice siempre la estratificación para la clasificación (especialmente en los casos desbalanceados)
    - División Train-test y Shuffle-split: útil para grandes conjuntos de datos 
- Selección de modelo
    - No agregue sobre los puntajes de test: aquellos que han visto los datos del test set
    - Utilice conjuntos de validación para elegir primero algoritmos / hiperparámetros
- Optimización
    - Grid Search: exhaustivo pero simple
    - Random Search: escala mejor

# Resumen II
* Los datos del mundo real a menudo están desbalanceados
* Los falsos positivos pueden ser mucho peores que los falsos negativos (o viceversa)
* Clasificación binaria
    - Seleccione métricas que puedan distinguir diferentes tipos de errores (precision, recall, f1-score, AUC,...)
    - Calibre los umbrales de decisión para la tarea en cuestión
    - Curvas de Precision-Recall y ROC: elija el mejor umbral o tome el área debajo de la curva
* Clasificación multiclase
    - Promedio Macro/Micro/weighted de puntajes por clase (one-vs-all)
* Regresión
    - (Root) mean squared/absolute error desde 0..Inf
    - R2 más fácil de interpretar
* Todas las medidas se pueden utilizar en cross-validation o grid/random search
* Clasificación Cost-sensitive: optimizar para cualquier matriz de costos o función de costos

# Regresión sobre Seúl Bike demand data set
'https://archive.ics.uci.edu/ml/machine-learning-databases/00560/SeoulBikeData.csv'

```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00560/SeoulBikeData.csv', 
                encoding='latin_1', parse_dates=['Date'])

data.head(10)
```

```python
# Codificamos con 1 cuando sea festivo
data['Holiday'] = np.where(data['Holiday']=='Holiday', 1,0)

# Codificamos con 1 cuando sea día de semana (lun-vie)
data['Functioning Day'] = np.where(data['Functioning Day']=='Yes', 1, 0)

# Codificando las variables categoricas con get_dummies
data = pd.get_dummies(data, drop_first=True)

# Codificamos la columna de Hora como sen y cos:
data['cos_hour'] = np.cos(2*np.pi*data['Hour']/23)
data['sin_hour'] = np.sin(2*np.pi*data['Hour']/23)

# Borramos la columna de la fecha y hora
data.drop(columns=['Date','Hour'], inplace = True)
# por simplicidad movemos la variable dependiente a la última posición
col2move = data.pop('Rented Bike Count')
data['Rented Bike Count'] = col2move

print(f'el tamaño del dataFrame es :{data.shape}')
data.head(10)
```

```python
from sklearn.model_selection import train_test_split
data_trainVal, data_test = train_test_split(data, train_size= 0.85, random_state = 0)
data_train, data_val = train_test_split(data_trainVal, train_size = 0.9, random_state = 0)

print(f'Tamaño conjunto de entrenamiento (data_train): {data_train.shape}')
print(f'Tamaño conjunto de validación (data_val): {data_val.shape}')
print(f'Tamaño conjunto de test (data_test): {data_test.shape}')
```

```python
## Pre-processing
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler

# standarizar aquellas variables que no poseen era categoricas o sin cos
# lista con los nombres de las columnas a transofrmar
columns2transform = list(data.columns[:-8])
columns2transform.append(data.columns[-1])


column_transformer = make_column_transformer((StandardScaler(), columns2transform), remainder = 'passthrough')

data_train_scaler = column_transformer.fit_transform(data_train)
data_val_scaler = column_transformer.transform(data_val)
data_test_scaler = column_transformer.transform(data_test)
```

```python
meanY = data_train['Rented Bike Count'].mean()
stdY = data_train['Rented Bike Count'].std()

X_train, y_train = data_train_scaler[:,:-1], data_train_scaler[:,-1]
X_val, y_val = data_val_scaler[:,:-1], data_val_scaler[:,-1]
X_test, y_test = data_test_scaler[:,:-1], data_test_scaler[:,-1]


print('(Train shape) X: {}, y: {}'.format(X_train.shape, y_train.shape))
print('(Validation shape) X: {}, y: {}'.format(X_val.shape, y_val.shape))
print('(Test shape) X: {}, y: {}'.format(X_test.shape, y_test.shape))

print('mean Y: {}, std y: {}'.format(meanY, stdY))
```

```python
# Construir modelo

## tensorflow-Keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras import metrics


model = Sequential(name='Regression')
model.add(layers.Dense(units=15,
                       activation='relu',
                       input_shape=(X_train.shape[1],),
                       name='Hidden_01'))

model.add(layers.Dense(units=23,
                       activation='relu',
                       name='Hidden_02'))

model.add(layers.Dense(units=27,
                       activation='relu',
                       name='Hidden_03'))

model.add(layers.Dense(units=1,
                       activation='linear',
                       name='Output'))

# compiler settings
model.compile(optimizer = 'adam',
              loss= 'mae',
              metrics=[metrics.MeanSquaredError(), metrics.MeanAbsolutePercentageError()])

# display model
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
estop = EarlyStopping(monitor='val_loss', patience=2, restore_best_weights=True)
```

```python
from time import time
start = time()
history = model.fit(x=X_train, y=y_train,
                    validation_data=(X_val, y_val),
                    epochs = 200,
                    batch_size=32,
                    callbacks=[estop])

print(f'Elapsed time [s]: {time()-start}')
```

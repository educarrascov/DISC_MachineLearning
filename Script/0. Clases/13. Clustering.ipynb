#### Load the CallCenterData dataset

```python
import pandas as pd
data = pd.read_csv('dataset/CallCenterData.csv')
```

#### Summary with descriptive statistics

```python
data.describe(include='all')
```

#### We are going to use only continuous attributes

```python
X=data[['Edad','HorasLogeadas','HorasHabladas','ContactosEfectivosPromedio','RegistrosTerminados']]
X.head(5)
```

#### Since the attributes are on different scales, it is convenient to standardize 

```python
from sklearn.preprocessing import StandardScaler
# Standardize the data so that they have mean 0 and variance 1
X_std = pd.DataFrame(StandardScaler().fit_transform(X),columns=X.columns)

#shows the first 5 rows of the table
X_std.head(5)
```

#K-means

#### Lets run k-means with k=2

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kfit = kmeans.fit(X_std)
identified_clusters = kfit.predict(X_std)
print(identified_clusters)
```

#### Why k = 2, how can we determine that? $\rightarrow$ elbow method

```python
import matplotlib.pyplot as plt

#the sum of the squared error for different values of k
SSE = []
for i in range(1, 11):
  km = KMeans(n_clusters=i)
  km.fit(X_std)
  SSE.append(km.inertia_) #nombre inertia = MSSE

# plot

plt.plot(range(1, 11), SSE,'-o', color='black')
plt.xlabel('N√∫mero of clusters')
plt.ylabel('SSE')
plt.show()
```

#### It seems that k = 3 is appropriate (k = 4 would also be ok)

```python
kmeans = KMeans(n_clusters=3)
kfit = kmeans.fit(X_std)
#the centroids
print(kfit.cluster_centers_,'\n')
#the labels
identified_clusters = kfit.predict(X_std)
print(identified_clusters)
```

#### Let's visualize using PCA





```python
from sklearn.decomposition import PCA

# Create a PCA instance: pca
pca = PCA()
principalComponents = pca.fit_transform(X_std)

# Plot the percentage of the variance that is explained by each component
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save the components in a dataframe
PCA_components = pd.DataFrame(principalComponents)
```

#### We will use the two main components to project our data from 5D to 2D and color according to the cluster label

```python
fig, ax = plt.subplots()
scatter = ax.scatter(PCA_components[0], PCA_components[1], alpha=.5, 
            c=identified_clusters)

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Clusters")
ax.add_artist(legend1)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

#  Density-Based Spatial Clustering of Applications with Noise (DBSCAN)

Original paper can be found [here](https://www.aaai.org/Papers/KDD/1996/KDD96-037.pdf)

#### We first need to determine an optimal epsilon (Eps) value
Method proposed [here](https://iopscience.iop.org/article/10.1088/1755-1315/31/1/012012/pdf)

```python
import numpy as np

from sklearn.neighbors import NearestNeighbors

#We will consider MinPts = 3
neigh = NearestNeighbors(n_neighbors=3)
nbrs = neigh.fit(X_std)
distances, indices = nbrs.kneighbors(X_std)

distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

We notice that for MinPts=3, Eps is between 0.8-1.0. We will consider Eps=0.9. 

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.9, min_samples=3)
db.fit(X_std)
print(db.labels_)
# los que presente -1, son los noise points
```

label -1 corresponds to noisy points.

```python
#PCA visualization

fig, ax = plt.subplots()
scatter = ax.scatter(PCA_components[0], PCA_components[1], alpha=.5, 
            c=db.labels_)

legend1 = ax.legend(*scatter.legend_elements(),
                    loc="upper right", title="Clusters")
ax.add_artist(legend1)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()

#los -1 corresponden a noise points
```
